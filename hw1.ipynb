```python
#2

# Check for duplicated rows considering all columns
duplicated_rows = merged_df[merged_df.duplicated(keep=False)]

num_duplicated = len(duplicated_rows)

# Display the number of duplicated rows
print("Number of duplicated rows:", num_duplicated , "\n")

# Display duplicated rows if there is at least 1.
if num_duplicated > 0:
    print("Duplicated rows:")
    print(duplicated_rows)

# Calculate mean, median, standard deviation, and other summary statistics
statistics = merged_df.describe()
statistics
```

**Set of Questions on SQL - using sqlite3 library!**

**Dataset Path: "Data/supermarket_sales.csv"**


The Supermarket Sales Dataset is a comprehensive dataset designed to provide insights into the sales performance of a supermarket.

We want to store this data in a relational database, with the following table schemas:

# Tables

## 1. Branches Table

| Column   | Type  | Key      | Description                              |
|----------|-------|----------|------------------------------------------|
| BranchID | INTEGER  | PRIMARY  | Unique identifier for each branch        |
| Branch | TEXT  |            |    Name of the supermarket branch        |
| City     | TEXT  |          | City where the branch is located         |



## 2. Products Table

| Column      | Type  | Key      | Description                               |
|-------------|-------|----------|-------------------------------------------|
| ProductID | INTEGER  | PRIMARY  | Unique identifier for each product line   |
| ProductLine | TEXT  |          | Category of the product |



## 3. Orders Table

| Column        | Type  | Key      | Description                                    |
|---------------|-------|----------|------------------------------------------------|
| OrderID       | TEXT  | PRIMARY  | Unique identifier for each order               |
| BranchID      | INTEGER  |   FK       | Identifier for the branch of the order    |
| ProductID   | INTEGER  |      FK    | Identifier linking to Products table ordered                      |
| Quantity      | INTEGER |        | Number of units ordered                        |
| UnitPrice   | REAL  |          | Price per unit of the product             |Total         | REAL  |          | Total sales value for the order               |
| OrderDate     | TEXT  |          | Date when the order was placed                |
| PaymentMethod | TEXT  |          | Method used for payment                       |
| CustomerType  | TEXT  |          | Type of customer (e.g., Member, Normal)     |
| Gender        | TEXT  |          | Gender of the customer                        |



Summary of Table Structures

* Branches Table contains information about different supermarket branches.

* Products Table details the various products available for purchase.
* Orders Table records the transactions made, linking products and branches to specific purchases.


# **Questions** **(35 Pts)**

1. Read the csv file and create this DB **(10 Pts)**
2. Write a query to find the top 3 product lines with the highest total quantity sold, using a LIMIT statement. **(4 Pts)**
3. Find the total revenue generated for each branch by product line **(4 Pts)**
4. Find the total sales for each branch **(3 Pts)**
5. Determine the number of orders placed by each gender **(3 Pts)**
6. List the product lines sold in each branch and the number of orders **(4 Pts)**
7. Find the highest quantity sold for a single order and the corresponding product line **(4 Pts)**
8. Check for duplicated invoices **(3 Pts)**


```python
#1


# Load the CSV file.
df = pd.read_csv('Data/supermarket_sales.csv')

# Connect to the SQLite database.
with sqlite3.connect('supermarket.db') as conn:
    cursor = conn.cursor()

    # Create the Branches table.
    df_branches = df[['Branch', 'City']].drop_duplicates().reset_index(drop=True)
    df_branches['BranchID'] = df_branches.index + 1

    # Reorder columns.
    df_branches = df_branches[['BranchID', 'Branch', 'City']]

    # Create the Branches table in the database with BranchID as the primary key.
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Branches (
        BranchID INTEGER PRIMARY KEY,
        Branch TEXT,
        City TEXT
    )
    ''')
    df_branches.to_sql('Branches', conn, if_exists='replace', index=False)

    # Create the Products table.
    df_products = df[['Product line']].drop_duplicates().reset_index(drop=True)
    df_products['ProductID'] = df_products.index + 1

    # Reorder columns.
    df_products = df_products[['ProductID', 'Product line']]

    # Create the Products table in the database with ProductID as the primary key.
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Products (
        ProductID INTEGER PRIMARY KEY,
        ProductLine TEXT
    )
    ''')
    df_products.to_sql('Products', conn, if_exists='replace', index=False)

    # Prepare order data by using Invoice ID as OrderID.
    df_orders = df[['Invoice ID', 'Branch', 'Product line', 'Quantity', 'Unit price', 'Date', 'Payment', 'Customer type', 'Gender']].reset_index(drop=True)
    df_orders.rename(columns={'Invoice ID': 'OrderID'}, inplace=True)


    # Copy the original OrderID column before dropping duplicates to analysis invoide id for last question.
    df_orders_copy = df_orders.copy()

    # Drop duplicate OrderID rows because OrderID is primary key for table.
    df_orders = df_orders.drop_duplicates(subset=['OrderID']).reset_index(drop=True)




    # Merge order data with BranchID and ProductID to link the data.
    df_orders = df_orders.merge(df_branches[['Branch', 'BranchID']], on='Branch')
    df_orders = df_orders.merge(df_products[['Product line', 'ProductID']], on='Product line')

    # Create the Orders table in the database.
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Orders (
        OrderID TEXT PRIMARY KEY,
        BranchID INTEGER,
        ProductID INTEGER,
        Quantity INTEGER,
        UnitPrice REAL,
        OrderDate TEXT,
        PaymentMethod TEXT,
        CustomerType TEXT,
        Gender TEXT,
        FOREIGN KEY (BranchID) REFERENCES Branches(BranchID),
        FOREIGN KEY (ProductID) REFERENCES Products(ProductID)
    )
    ''')

    # Insert the processed order data into the Orders table.
    df_orders = df_orders[['OrderID', 'BranchID', 'ProductID', 'Quantity', 'Unit price', 'Date', 'Payment', 'Customer type', 'Gender']]
    df_orders.to_sql('Orders', conn, if_exists='replace', index=False)

    # Display all tables in the database.
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()
    print("Current Tables:", tables)
```

```python
#2

# Finding the top 3 product lines with the highest total quantity sold
cursor.execute('''
SELECT p."Product line", SUM(o.Quantity)
FROM Products p, Orders o
WHERE o.ProductID = p.ProductID
GROUP BY p."Product line"
ORDER BY SUM(o.Quantity) DESC
LIMIT 3;
''')

# Fetch all results from the query
all = cursor.fetchall()

# Convert the result to a df for better displaying
df = pd.DataFrame(all, columns=['ProductLine', 'Highest Quantity'])

# Display the DataFrame as a table
df
```

```python
#3

# Calculating revenue by multiplying Quantity and Unit price for each product line by brach.
cursor.execute('''
SELECT
    b.Branch,
    p."Product line",
    SUM(o.Quantity * o."Unit price") AS Revenue
FROM
    Branches as b, Products as p, Orders as o
WHERE
    o.BranchID = b.BranchID AND
    o.ProductID = p.ProductID
GROUP BY
    b.Branch, p."Product line"
''')

# Fetch all results from the query
all = cursor.fetchall()

# Convert the result to a df for better displaying
df = pd.DataFrame(all, columns=['Branch', 'Product line', 'Revenue'])

# Display the DataFrame as a table
df
```

```python
#4

# Summing up total sales (Quantity) for each branch
cursor.execute('''
SELECT b.Branch, SUM(o.Quantity) AS TotalSales
FROM Orders as o
JOIN Branches as b
ON o.BranchID = b.BranchID
GROUP BY b.Branch
''')

# Fetch all results from the query
all = cursor.fetchall()

# Convert the result to a df
df = pd.DataFrame(all, columns=['Branch', 'TotalSales'])

# Display the DataFrame as a table
df
```

```python
#5

#Determine the number of orders placed by each gender
cursor.execute('''
SELECT Gender, COUNT(*)
FROM Orders
GROUP BY Gender
''')

# Fetch all results from the query
all = cursor.fetchall()

# Convert the result to a df
df = pd.DataFrame(all, columns=['Gender', 'Number of Orders'])

# Display the DataFrame as a table
df
```

```python
#6

# List the product lines sold in each branch and the number of orders
cursor.execute('''
SELECT b.Branch, p."Product line", COUNT(*)
FROM Orders as o, Branches as b, Products as p
WHERE o.BranchID = b.BranchID AND
o.ProductID = p.ProductID
GROUP BY b.Branch, p."Product line"
''')


# Fetch all results
all = cursor.fetchall()

# Convert the result to a df
df = pd.DataFrame(all, columns=['Brach', 'ProductLine', "NumberOfOrders"])

# Display the DataFrame as a table
df
```

```python
#7

#Finding the highest quantity sold for a single order and the corresponding product line
#  by sorting the descending order and select the top 1 product.
cursor.execute('''
SELECT Quantity, p."Product line"
FROM Orders as o
JOIN Products as p ON o.ProductID = p.ProductID
ORDER BY Quantity DESC
LIMIT 1
''')

# Fetch all results
all = cursor.fetchall()

# Display the result
print(all)
```

```python
#8

# Group the results by OrderID and find those with a count greater than 1 (duplicates)
cursor.execute('''
SELECT OrderID
FROM Orders
GROUP BY OrderID
HAVING COUNT(*) > 1;
''')

# Fetch all the results from the executed query
all = cursor.fetchall()

# Check if there are any duplicate OrderIDs
is_duplicate = len(all) > 0

# If duplicates exist print duplicates.
if is_duplicate:
    print("There are some duplicated invoices:")
    print("Duplicated ID's")
    for it in all:
        print(it)
else:
    print("No duplicate invoices found.")
```

```python
#8
# Unlike the previous OrderID primary key duplicate analysis,
# the following block checks for duplicates in Invoice ID without applying drop.duplicate.



# Save df_orders_copy as a temporary table in the SQLite database
df_orders_copy.to_sql('OrdersCopy', conn, if_exists='replace', index=False)

# Execute an SQL query to select duplicate OrderID from the temporary OrdersCopy table
cursor.execute('''
SELECT OrderID
FROM OrdersCopy
GROUP BY OrderID
HAVING COUNT(*) > 1;
''')

# Fetch all the results from the executed query
duplicate_copy_orders = cursor.fetchall()

# Check if there are any duplicate OrderIDs in the copied dataframe
is_duplicate_copy = len(duplicate_copy_orders) > 0

# If duplicates exist, print duplicates.
if is_duplicate_copy:
    print("There are some duplicated invoices in the copied dataframe (df_orders_copy):")
    print("Duplicated ID's")
    for it in duplicate_copy_orders:
        print(it)
else:
    print("No duplicate invoices found in the copied dataframe (df_orders_copy).")
```
